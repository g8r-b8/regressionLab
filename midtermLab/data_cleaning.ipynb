```python
replacements_NU = {
    'No':0,
    'YES':1,
    '.':'NaN'
}
data['NU'].replace(replacements_NU, inplace=True)
data
```

```python
replacements_G = {
    'M':0,
    'F':1,
    '.':'NaN'
}
data['CG'].replace(replacements_G, inplace=True)
data
```
