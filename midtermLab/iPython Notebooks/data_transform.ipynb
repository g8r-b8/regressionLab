```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression

data = pd.read_csv('height_data_cleaned.csv') 
```

```python
data
```

```python
##get the values as
#x, y, X_new

X=data[['FH','FR','MH','MR','CG','NU']]
Y=data[['CH']]
```

```python
XX=X.multiply(X)
```

```python

XX.columns = [str(col) + '_Squared' for col in XX.columns]
```

```python
X2=pd.concat([X,XX],axis=1)
```

```python
X2
```

```python
xDif=X['MH'].subtract(X['FH'])
```

```python
type(xDif)
```

```python
xDist=xDif.abs()
```

```python
xTallerParent=xDif.divide(xDist)
```

```python
Xtemp=pd.concat([xDist,xTallerParent],axis=1)
```

```python
Xtemp
```

```python
Xtemp.columns=["Distance","TallerParent"]
```

```python
Xtemp
Xtemp_S=Xtemp.multiply(Xtemp)

Xtemp_S.columns = [str(col) + '_Squared' for col in Xtemp_S.columns]
```

```python
X3=pd.concat([X2,Xtemp,Xtemp_S],axis=1)
```

```python
finData=pd.concat([X3,Y],axis=1)
```

```python
finData.to_csv('./height_data_transformed.csv', index = False)
```
