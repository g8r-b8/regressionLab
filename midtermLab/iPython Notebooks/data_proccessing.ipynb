```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import decomposition
from sklearn import ensemble
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectFromModel
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd 

```

```python
data = pd.read_csv('height_data_transformed_no_squares.csv') 
data
```

```python
X=data[['FH','FR','MH','MR','CG','NU','Distance','TallerParent']]
y=data[['CH']]
X
```

```python
def split(df, headSize) :
    hd = df.head(headSize)
    tl = df.tail(len(df)-headSize)
    return hd, tl

```

```python
X_training, X_test = split(X, 260)
Y_training, Y_test = split(y, 260)
X_training
X_test
```

```python
from sklearn.linear_model import SGDRegressor

for style, width, degree in (   ("b--", 2, 2), ("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False) 
    std_scaler = StandardScaler()
    lin_reg = SGDRegressor()  
    pca = decomposition.PCA(n_components=8)
    
    polynomial_regression = Pipeline([
            ("poly_features", polybig_features),
            ("std_scaler", std_scaler),
            ("pca",pca),
            ("lin_reg", lin_reg),
            
           
        ])
    Xnew=pca.fit_transform(X_training)
    Xtnew=pca.transform(X_test)
    #cv = StratifiedKFold( n_splits=5, shuffle=True, random_state=2016)
    #score = cross_val_score(polynomial_regression,X_training,Y_training,cv=cv)
    #print(score)   
  
    print(pca.explained_variance_ratio_)

    polynomial_regression.fit(Xnew, Y_training)
    
    y_newbig = polynomial_regression.predict(Xtnew)
    #print(lin_reg.coef_)
    print(polynomial_regression.score(Xtnew,Y_test))
    #plt.plot(X_test, y_newbig, style, label=str(degree), linewidth=width)
```
